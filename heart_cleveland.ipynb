```python
# Calcular Z-score para todo el DataFrame
z_scores = np.abs(stats.zscore(Datos))
```

```python
# Definir umbral para considerar un valor como atípico (3 es estándar)
umbral = 3

# Crear un DataFrame booleano: True si el valor es un outlier
outliers_mask = (z_scores > umbral)

# Filas con al menos un outlier
datos_atipicos = Datos[outliers_mask.any(axis=1)]

# Mostrar los datos atípicos
print("Datos atípicos encontrados:")
print(datos_atipicos)
```
